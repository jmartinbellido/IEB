```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import date
import datetime
import numpy as np
```

##### EX 3: Bar Charts
---
Build a bar chart using Matplotlib to visualize average GDP per capita by continent. Highlight Europe on a different color.

> https://data-wizards.s3.amazonaws.com/datasets/dataset_na_who.csv

![title](https://data-wizards.s3.amazonaws.com/img/matplotlib/EX5.png)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import date
import datetime
import numpy as np
```

```python
%matplotlib inline
```

```python
df_who = pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/dataset_na_who.csv")
df_who.dtypes
```

```python
df = df_who.groupby("Continent")\
    .agg({'Gross national income per capita (PPP international $)':'sum'})\
    .rename({'Gross national income per capita (PPP international $)':'GDP_per_capita'},axis=1)\
    .sort_values('GDP_per_capita',ascending=False)

df = df.reset_index()

fig, ax = plt.subplots(figsize = (12,8))

ax.set_title("GDP Per Capita by Continent\nWorld Health Organization")
ax.bar(
    df['Continent'],
    df['GDP_per_capita'],
    width=0.75, 
    edgecolor="white", 
    linewidth=1,
    facecolor='grey'
)

ax.bar(
    df[df.Continent =='Europe']['Continent'],
    df[df.Continent =='Europe']['GDP_per_capita'],
    width=0.75, 
    edgecolor="white", 
    linewidth=1,
    facecolor='orange'
)

ax.set_xlabel("Continent")
ax.set_ylabel("GDP Per Capita")
```

##### EX 5: Time Series
---
The dataset below contains historical data for an ecommerce. Build a time series chart to visualize weekly orders and visits during H1 2017 (see image below).

> https://data-wizards.s3.amazonaws.com/datasets/ga_merchant_store.csv

![title](https://data-wizards.s3.amazonaws.com/img/matplotlib/EX3.png)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import date
import datetime
import numpy as np
```

```python
%matplotlib inline
```

```python
df_google_analytics = pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/ga_merchant_store.csv")
df_google_analytics.dtypes
```

```python
df_google_analytics['date'] = pd.to_datetime(df_google_analytics['date'])
df_google_analytics.dtypes
```

```python
df_google_analytics['week'] = df_google_analytics['date'].dt.isocalendar().week
df_google_analytics['month'] = df_google_analytics['date'].dt.month
df_google_analytics['year'] = df_google_analytics['date'].dt.year
df_google_analytics.head()
```

```python
cond_1 = df_google_analytics.year == 2017
cond_2 = df_google_analytics.month < 7
cond_3 = df_google_analytics.week==52

df = df_google_analytics[cond_1 & cond_2 & -cond_3].groupby('week').agg({'transactions':'sum','visits':'sum'})
df = df.reset_index()
df
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,8))

ax[0].set_title("H1 2017\nWeekly Transactions")
ax[0].set_ylabel("Orders")
ax[0].plot(
    df['week'],
    df['transactions'],
    color = 'red',
    linewidth=1.5
)

ax[1].set_title("Weekly Visits")
ax[1].set_ylabel("Visits (K)")
ax[1].plot(
    df['week'],
    df['visits']/1000,
    color = 'blue',
    linewidth=1.5
)

ax[1].set_xlabel("Week Number")

ax[0].set_xticks(np.arange(df['week'].min(), df['week'].max(), step=1))
ax[1].set_xticks(np.arange(df['week'].min(), df['week'].max(), step=1))
```

##### EX 6: Time Series
---
Repeat exercise 5, this time include bar charts to show top channels and sources in terms of total visits during Hi 2017.

> https://data-wizards.s3.amazonaws.com/datasets/ga_merchant_store.csv

![title](https://data-wizards.s3.amazonaws.com/img/matplotlib/EX4.png)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import date
import datetime
import numpy as np
```

```python
%matplotlib inline
```

```python
df_google_analytics = pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/ga_merchant_store.csv")
df_google_analytics.dtypes
```

```python
df_google_analytics['date'] = pd.to_datetime(df_google_analytics['date'])
df_google_analytics.dtypes
```

```python
df_google_analytics['week'] = df_google_analytics['date'].dt.isocalendar().week
df_google_analytics['month'] = df_google_analytics['date'].dt.month
df_google_analytics['year'] = df_google_analytics['date'].dt.year
df_google_analytics.head()
```

```python
cond_1 = df_google_analytics.year == 2017
cond_2 = df_google_analytics.month < 7
cond_3 = df_google_analytics.week==52

fig, ax = plt.subplots(2,2, figsize = (15,10))

fig.tight_layout(pad=9)

# charts 1 & 2

df = df_google_analytics[cond_1 & cond_2 & -cond_3].groupby('week').agg({'transactions':'sum','visits':'sum'})
df = df.reset_index()

ax[0,0].set_title("H1 2017\nWeekly Transactions")
ax[0,0].set_ylabel("Orders")
ax[0,0].plot(
    df['week'],
    df['transactions'],
    color = 'red',
    linewidth=1.5
)

ax[1,0].set_title("H1 2017\nWeekly Visits")
ax[1,0].set_ylabel("Visits (K)")
ax[1,0].plot(
    df['week'],
    df['visits']/1000,
    color = 'blue',
    linewidth=1.5
)

ax[1,0].set_xlabel("Week Number")

ax[0,0].set_xticks(np.arange(df['week'].min(), df['week'].max(), step=2))
ax[1,0].set_xticks(np.arange(df['week'].min(), df['week'].max(), step=2))

# charts 3

df = df_google_analytics[cond_1 & cond_2 & -cond_3]\
    .groupby('channelGrouping')\
    .agg({'visits':'sum'})\
    .sort_values('visits')
df = df.reset_index()

ax[0,1].set_title("H1 2017\nTotal Visits by Channel")
ax[0,1].barh(
    df['channelGrouping'],
    df['visits']/1000,
    #width=0.75, 
    edgecolor="white", 
    linewidth=1,
    facecolor='grey'
)
ax[0,1].set_xlabel("Visits (K)")

# chart 4

df = df_google_analytics[cond_1 & cond_2 & -cond_3]\
    .groupby('source')\
    .agg({'visits':'sum'})\
    .sort_values('visits',ascending=False)\
    .head(7)\
    .sort_values('visits')
df = df.reset_index()

ax[1,1].set_title("H1 2017\nTotal Visits by Source")
ax[1,1].barh(
    df['source'],
    df['visits']/1000,
    #width=0.75, 
    edgecolor="white", 
    linewidth=1,
    facecolor='grey'
)
ax[1,1].set_xlabel("Visits (K)")
```
